```python
#loading the dataset
import pandas as pd
df=pd.read_csv('/all_kindle_review.csv')
```

```python
df=df[['reviewText','rating']]
df.head()
```

```python
#Cleaning and preprocessing the data
#tagging postive and negative reviews as 1 and 0 respectively
```

```python
df['rating']=df['rating'].apply(lambda x:1 if x>2 else 0)
```

```python
df['rating'].value_counts()
```

```python
#lowering all the text
df['reviewText']=df['reviewText'].str.lower()
```

```python
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
import re
from bs4 import BeautifulSoup
```

```python
#removing special keywords
df['reviewText']=df['reviewText'].apply(lambda x:re.sub('[^a-z A-Z 0-9]+', '',x))
#removing stopwords
df['reviewText']=df['reviewText'].apply(lambda x:" ".join([ y for y in x.split() if y not in stopwords.words('english')]))
#removing urls
df['reviewText']=df['reviewText'].apply(lambda x: re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , str(x)))
#removing html tags
df['reviewText']=df['reviewText'].apply(lambda x: BeautifulSoup(x,'lxml').get_text())
#removing any additional spaces
df['reviewText']=df['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
df.tail()
```

```python
#LEMMATISING
```

```python
##lemmatising the text
from nltk.stem import WordNetLemmatizer
lemmatiser=WordNetLemmatizer()
```

```python
def lemmatise_words(text):
  return " ".join([lemmatiser.lemmatize(word) for word in text.split()])
```

```python
nltk.download('wordnet')
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:lemmatise_words(x))
```

```python
df.head()
```

```python
#DOING TRAIN TEST SPLIT
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df['reviewText'],df['rating'],test_size=0.20,random_state=42)
```

```python
#APPLYING WORD EMBEDDINGS
```

```python
#Applying Bag of Words Word Embdedding Method
from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer()
X_train_bow=bow.fit_transform(X_train).toarray()
X_test_bow=bow.transform(X_test).toarray()
```

```python
#Applying  Word EmbdedTF-IDF Embedding Method
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=CountVectorizer()
X_train_tfidf=tfidf.fit_transform(X_train).toarray()
X_test_tfidf=tfidf.transform(X_test).toarray()
```

```python
#TRAINING OUR MODEL
from sklearn.linear_model import LogisticRegression
nb_bow_model=LogisticRegression().fit(X_train_bow,y_train)
nb_tfidf_model=LogisticRegression().fit(X_train_tfidf,y_train)
```

```python
#PERORMANCE PARAMETERS
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
y_pred_bow=nb_bow_model.predict(X_test_bow)
```

```python
y_pred_tfidf=nb_tfidf_model.predict(X_test_tfidf)
```

```python
#bow confusion matrix
```

```python
confusion_matrix(y_test,y_pred_bow)
```

```python
#bow accuracy
print("Bag of Words accuracy is :",accuracy_score(y_test,y_pred_bow))
```

```python
print(classification_report(y_test,y_pred_bow))
```

```python
#TFIDF confusion matrix
confusion_matrix(y_test,y_pred_tfidf)
```

```python
#TFIDF Accuracy
print("TFIDF  accuracy is :",accuracy_score(y_test,y_pred_tfidf))
```

```python
print(classification_report(y_test,y_pred_tfidf))
```
